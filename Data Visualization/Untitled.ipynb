## Import Libraries

```python
# import Libraries

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("tips_data.csv")
df.head()
```

```python
df.info()
```

## DATA VISUALIZATION

### Scatter Plot

```python

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['total_bill'], df['tip'])
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()
```

### Pie Chart

```python
# Pie chart
plt.figure(figsize=(8, 6))
df['sex'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Pie Chart of Sex Distribution')
plt.axis('equal')
plt.show()
```

### Bar Chart

```python
# Bar chart
plt.figure(figsize=(8, 6))
df.groupby('day')['total_bill'].mean().plot(kind='bar')
plt.xlabel('Day')
plt.ylabel('Average Total Bill')
plt.title('Bar Chart of Average Total Bill by Day')
plt.show()
```

### Subplots

```python


# Subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Scatter plot
axes[0, 0].scatter(df['total_bill'], df['tip'])
axes[0, 0].set_xlabel('Total Bill')
axes[0, 0].set_ylabel('Tip')
axes[0, 0].set_title('Scatter Plot of Total Bill vs Tip')

# Pie chart
axes[0, 1].pie(df['sex'].value_counts(), autopct='%1.1f%%')
axes[0, 1].set_title('Pie Chart of Sex Distribution')
axes[0, 1].axis('equal')


# Bar chart
axes[1, 0].bar(df['day'], df['total_bill'])
axes[1, 0].set_xlabel('Day')
axes[1, 0].set_ylabel('Total Bill')
axes[1, 0].set_title('Bar Chart of Total Bill by Day')

# Histogram
axes[1, 1].hist(df['total_bill'], bins=10)
axes[1, 1].set_xlabel('Total Bill')
axes[1, 1].set_ylabel('Count')
axes[1, 1].set_title('Histogram of Total Bill')

plt.tight_layout()
plt.show()
```

### Heatmap

```python
# Select only the numeric columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Calculate the correlation matrix using the numeric columns
corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='YlOrRd', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.show()
```
